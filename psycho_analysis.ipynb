```python
i = psycho_df["EEG Power Bands"]
for i in i:
   v = eval(i)
   
```

```python
Delta = []
Alpha = []
Beta = []

for val in psycho_df["EEG Power Bands"]:
    converted_format = eval(val)
    Delta.append(converted_format[0])
    Alpha.append(converted_format[1])
    Beta.append(converted_format[2])
    
psycho_df["Delta"] = Delta
psycho_df["Alpha"] = Alpha
psycho_df["Beta"] = Beta
```

```python
psycho_df.drop(columns = ["Time","EEG Power Bands"],axis = 1,inplace = True)
```

```python
psycho_df = psycho_df.round(3)
```

```python
#psycho_df.select_dtypes(exclude = ["int","float"])
from sklearn.preprocessing import LabelEncoder
#psycho_df = psycho_df.select_dtypes(include=["number","object"])

column_features = psycho_df.columns.tolist()
for features in column_features:
    encode = LabelEncoder()
    if psycho_df[features].dtype in ["int64","float64"]:
         continue
    else:
        psycho_df[features] = encode.fit_transform(psycho_df[features])
```

```python
psycho_df = pd.concat([psycho_df.iloc[:,:7],psycho_df.iloc[:,8:],psycho_df.iloc[:,7]],axis = 1)
```

```python
psycho_df
```

```python
X = psycho_df.iloc[:,:-1]
Y = psycho_df.iloc[:,-1]
from sklearn.decomposition import PCA

pca = PCA(n_components=18)  # Reduce to 5 principal components
X_reduced = pca.fit_transform(X)
```

```python
# port models
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix

x_train,x_test,y_train,y_test = train_test_split(X_reduced,Y,test_size = 0.2,random_state = 42)
scaler = StandardScaler()
new_x_train = scaler.fit_transform(x_train)
new_x_test = scaler.transform(x_test)

ensemble = RandomForestClassifier(n_estimators = 5,criterion = 'entropy')
ensemble.fit(new_x_train,y_train)
ensemble_pred = ensemble.predict(new_x_test)
# print(ensemble_pred[:5])
# print(y_test[:5])
print(classification_report(y_test,ensemble_pred))
print(confusion_matrix(y_test,ensemble_pred))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel='rbf',decision_function_shape='ovo')
svm.fit(new_x_train,y_train)
svm_pred = svm.predict(new_x_test)
print(svm_pred[:5])
print(classification_report(y_test,svm_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
sv = KNeighborsClassifier(n_neighbors = 5)
sv.fit(new_x_train,y_train)
sv_pred = sv.predict(new_x_test)
print(sv_pred[:5])
print(y_test[:5])
print(classification_report(y_test,sv_pred))
```

```python
# Create neural network
import tensorflow as tf 

# seed
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(50,activation = "relu"),
    tf.keras.layers.Dense(50,activation = "relu"),
    tf.keras.layers.Dense(50,activation = "relu"),
    tf.keras.layers.Dense(30,activation = "relu"),
    tf.keras.layers.Dense(30,activation = "relu"),
    tf.keras.layers.Dense(1,activation = "softmax")
    
])

model.compile(
    loss = tf.keras.losses.CategoricalCrossentropy(),
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["accuracy"]
)

history = model.fit(new_x_train,y_train,epochs = 50)

```

```python
from scipy.stats import pearsonr

# Sample data
heart_rate = psycho_df["Mood State"]
psychological_state = psycho_df["Psychological State"]

# Calculate correlation and p-value
correlation, p_value = pearsonr(heart_rate, psychological_state)

print(f"Correlation: {correlation}")
print(f"P-value: {p_value}")

# Interpretation of p-value
if p_value < 0.05:
    print("The correlation is statistically significant (p < 0.05).")
else:
    print("The correlation is not statistically significant (p >= 0.05).")
```
